```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
import plotly.express as px



```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/Ramzy75/projet_perso/master/netflix_titles_nov_2019.csv")

display(dataset)
```

```python
def label_dataset(d):
    clmns = ['P_{}'.format(i) for i in range(d.shape[1])]
    d.columns = clmns
    
    return (d, clmns)
    
# dataset["First director"]= new[0]
new, new_director = label_dataset(dataset["director"].str.split(",", expand = True))
new['show_id'] = dataset['show_id']
display(new)


new1, new_cast = label_dataset(dataset["cast"].str.split(",", expand = True))
new1['show_id'] = dataset['show_id']
display(new1)


new2, new_country = label_dataset(dataset["country"].str.split(",", expand = True))
new2['show_id'] = dataset['show_id'] 
display(new2)

new3, new_listed_in = label_dataset(dataset["listed_in"].str.split(",",expand = True))
new3['show_id'] = dataset['show_id'] 
display(new3)



```

```python
#faire une fonction 

df_melt1 = pd.melt(new,id_vars=['show_id'], value_vars = new_director,value_name = 'director').dropna().drop (columns = ['variable'])
print(df_melt1.shape)
display(df_melt1)
#display(df_melt1[df['variable'] == 'P_1'])
df_melt1.to_csv('/home/wamont1-4/Bureau/3WA/pythonramzi/projet02/data_prep/df_melt1_new_director.csv', index = False)

df_melt2 = pd.melt(new1,id_vars=['show_id'], value_vars = new_cast).dropna()
print(df_melt2.shape)
display(df_melt2)
df_melt2.to_csv('/home/wamont1-4/Bureau/3WA/pythonramzi/projet02/data_prep/df_melt2_new_cast.csv', index = False)

df_melt3 = pd.melt(new2,id_vars=['show_id'], value_vars = new_country).dropna()
print(df_melt3.shape)
display(df_melt3)
df_melt3.to_csv('/home/wamont1-4/Bureau/3WA/pythonramzi/projet02/data_prep/df_melt3_new_country.csv', index = False)


df_melt4 = pd.melt(new3,id_vars=['show_id'], value_vars = new_listed_in).dropna()
print(df_melt4.shape)
display(df_melt4)
df_melt4.to_csv('/home/wamont1-4/Bureau/3WA/pythonramzi/projet02/data_prep/df_melt4_new_listed_in.csv', index = False)
df_melt1
```
