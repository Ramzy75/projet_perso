# 1.2 Mise en forme de la colonne date_added

```python
def cut_date_added(data, col_year= 'year_added', col_month= 'month_added', clean= False):
    data["date_added"] = pd.to_datetime(data['date_added'])
    #display(data["date_added"])
    data[col_year] = data['date_added'].dt.year
    #display (data['year_added'])
    data[col_month] = data['date_added'].dt.month
    #display(data['month_added'])
    if clean == True:
        data = data.drop(columns= ['date_added'])
    
    return data

def fill_missing_date_added(data, col_year= 'year_added', from_col= 'release_year'):
    data[col_year] = np.where (data[col_year].isna(), data[from_col], data[col_year]).astype('int32')
    #display(data)
    
    return data


col_annee = 'year_available'
col_month = 'month_available'

df = cut_date_added(dataset, col_year= col_annee, col_month= col_month, clean= True)
df = fill_missing_date_added(df, col_year= col_annee)

#df = fill_missing_date_added(cut_date_added(dataset, col_year= col_annee), col_year= col_annee)
#display(df)
#display(df[df[col_annee].isna()])


df.to_csv(path + 'dataprep_newdatatime.csv', index = False)
df.head()


```

# 1.3 Mise en forme de la colonne duration

```python
def cut_duration (data, col_season='season', col_duration='duration'):

    data['season_count'] = data[~data.duration.isna()].apply(lambda x : x[col_duration].split(" ")[0] if "Season" in x[col_duration] else "" , axis = 1)
    data['duration'] = data[~data.duration.isna()].apply(lambda x : x[col_duration].split(" ")[0] if "Season" not in x[col_duration] else "", axis = 1)
    data.to_csv(path + 'dataprep_newduration.csv', index = False)
    
    return data
df = cut_duration(df, col_season='season', col_duration='duration')
df.head(50)
```

# 1.4 Mise en forme de la colonne : director, cast, country, listed_in

```python
def label_data(source, column, on, sep= ','):
    d = source[column].str.split(sep, expand = True)
    clmns = ['P_{}'.format(i) for i in range(d.shape[1])]
    d.columns = clmns
    d[on] = source[on]
    return (d, clmns)
    

new, new_director = label_data(df, 'director', 'show_id')
display(new)

new1, new_cast = label_data(df, "cast", 'show_id')
display(new1)

new2, new_country = label_data(df, "country", 'show_id')
display(new2)

new3, new_listed_in = label_data(df, "listed_in", 'show_id')
display(new3)


```

```python
def df_melt (data, pivot, melted, col_name):
     return pd.melt(data, id_vars=[pivot], value_vars = melted, value_name = col_name).dropna().drop (columns = ['variable'])
    

df_melt1 = df_melt(new, 'show_id', new_director, 'director')

df_melt2 = df_melt(new1,'show_id', new_cast,'cast')
display(df_melt2[df_melt2['show_id'] == 81193313])

df_melt3 = df_melt(new2,'show_id', new_country,'country')

df_melt4 = df_melt(new3,'show_id', new_listed_in,'listed_in')



```

```python
melt_list = [df_melt1,df_melt2,df_melt3,df_melt4]

def df_merge (df, melt_list, on, how ='left'):
    for dfmelt in melt_list:
        df = pd.merge(left= df,right = dfmelt, on= on, how= how)
    return df.drop(columns = ['cast_x', 'director_x', 'listed_in_x','country_x'], axis = 1)

df = df_merge (df, melt_list, 'show_id')

df.to_csv(path + 'dataprep_final_df.csv', index = False)
display(df)
```
